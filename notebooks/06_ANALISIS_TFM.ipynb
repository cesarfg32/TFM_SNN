# Análisis de resultados TFM SNN + CL

Este notebook construye las tablas y figuras clave para la memoria a partir de
los resúmenes generados por `05_RESULTS.ipynb`.

Flujo asumido:

1. Has ejecutado `05_RESULTS.ipynb` para el preset `std`
   → carpeta `paperset_std_rate_YYYY-MM-DD_HHMM`.
2. Has ejecutado `05_RESULTS.ipynb` para el preset `accurate`
   → carpeta `paperset_accurate_rate_YYYY-MM-DD_HHMM`.

Este notebook **no recalcula nada**: solo lee de esas carpetas y, cuando
necesita curvas por época, baja a los `run_dir` originales.

```python
from pathlib import Path
from typing import Optional, Dict, List

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------------------
# 0. CONFIGURA AQUÍ LAS CARPETAS SUMMARY QUE QUIERES USAR
# ----------------------------------------------------------------------

BASE_SUMMARY = Path("../outputs/summary")

STD_DIR = BASE_SUMMARY / "paperset_std_rate_2025-12-11_2009"
ACC_DIR = BASE_SUMMARY / "paperset_accurate_rate_2025-12-11_2035"
# Ajusta los nombres si has vuelto a ejecutar 05_RESULTS.ipynb con otra fecha/hora.

# OUT raíz de runs individuales (se deduce de STD_DIR → .../outputs/summary/...)
OUTPUTS_ROOT = STD_DIR.parent.parent  # ../outputs

print("[INFO] STD_DIR:", STD_DIR)
print("[INFO] ACC_DIR:", ACC_DIR)
print("[INFO] OUTPUTS_ROOT:", OUTPUTS_ROOT)

# ----------------------------------------------------------------------
# 1. Helpers básicos
# ----------------------------------------------------------------------

def load_csv(path: Path) -> pd.DataFrame:
    if not path.exists():
        raise FileNotFoundError(f"No existe: {path}")
    return pd.read_csv(path)

def ensure_mae_aliases(df: pd.DataFrame) -> pd.DataFrame:
    """
    Crea alias c1_final_mae/c2_final_mae si solo existen las columnas
    circuitoX_final_mae.
    """
    df = df.copy()
    if "c1_final_mae" not in df.columns and "circuito1_final_mae" in df.columns:
        df["c1_final_mae"] = df["circuito1_final_mae"]
    if "c2_final_mae" not in df.columns and "circuito2_final_mae" in df.columns:
        df["c2_final_mae"] = df["circuito2_final_mae"]
    return df


def canonical_method(m: str) -> str:
    s = (m or "").lower()
    if s.startswith("ewc"):
        return "ewc"
    if s.startswith("as-snn"):
        return "as-snn"
    if s.startswith("sa-snn"):
        return "sa-snn"
    if s.startswith("sca-snn"):
        return "sca-snn"
    if s.startswith("rehearsal"):
        return "rehearsal"
    if s.startswith("naive"):
        return "naive"
    if s == "std":
        return "std"
    return s or "unknown"


# ----------------------------------------------------------------------
# 2. Carga de resúmenes (std + accurate)
# ----------------------------------------------------------------------

# Ficheros principales de cada summary
STD_RESULTS = STD_DIR / "results_table_fromdisk.csv"
ACC_RESULTS = ACC_DIR / "results_table_fromdisk.csv"

STD_WINNERS = STD_DIR / "winners_per_methodbase.csv"
ACC_WINNERS = ACC_DIR / "winners_per_methodbase.csv"

STD_SELECTION = STD_DIR / "selection_table.csv"
ACC_SELECTION = ACC_DIR / "selection_table.csv"

STD_PARETO = STD_DIR / "pareto.csv"
ACC_PARETO = ACC_DIR / "pareto.csv"

df_std_all = ensure_mae_aliases(load_csv(STD_RESULTS))
df_acc_all = ensure_mae_aliases(load_csv(ACC_RESULTS))

print("[INFO] df_std_all:", df_std_all.shape)
print("[INFO] df_acc_all:", df_acc_all.shape)

# winners_per_methodbase ya están filtrados por comparabilidad (vienen de selection_table)
winners_std = ensure_mae_aliases(load_csv(STD_WINNERS))
winners_acc = ensure_mae_aliases(load_csv(ACC_WINNERS))

# Aseguramos method_base normalizada (por si acaso)
for df_w in (winners_std, winners_acc):
    if "method_base" in df_w.columns:
        df_w["method_base"] = df_w["method_base"].astype(str).apply(canonical_method)
    else:
        df_w["method_base"] = df_w["method"].astype(str).apply(canonical_method)

print("[INFO] winners_std métodos:", sorted(winners_std["method_base"].unique()))
print("[INFO] winners_acc métodos:", sorted(winners_acc["method_base"].unique()))

# ----------------------------------------------------------------------
# 3. Tabla resumen: ganadores por método (std + accurate)
# ----------------------------------------------------------------------

cols_base = [
    "preset",
    "method_base",
    "run_dir",
    "c1_final_mae",
    "c2_final_mae",
    "avg_forget_rel",
    "emissions_kg",
]

if "score_comp" in winners_std.columns or "score_comp" in winners_acc.columns:
    cols_base.append("score_comp")

winners_all = pd.concat([winners_std, winners_acc], ignore_index=True)

# Normaliza preset (por si viene vacío en winners_per_methodbase)
if "preset" not in winners_all.columns or winners_all["preset"].isna().all():
    # Los de STD_DIR → 'std'; los de ACC_DIR → 'accurate'
    # Nos apoyamos en que STD_WINNERS y ACC_WINNERS se concatenan en ese orden
    n_std = len(winners_std)
    winners_all["preset"] = ["std"] * n_std + ["accurate"] * len(winners_acc)

cols_present = [c for c in cols_base if c in winners_all.columns]
winners_compact = winners_all[cols_present].copy()

winners_compact = winners_compact.sort_values(["method_base", "preset"])
display(winners_compact)

# Export para usar desde LaTeX o revisar con calma
out_tabla = STD_DIR.parent / "tabla_winners_por_metodo_std_acc.csv"
winners_compact.to_csv(out_tabla, index=False)
print("[OK] tabla_winners_por_metodo_std_acc.csv →", out_tabla)

# ----------------------------------------------------------------------
# 4. Figuras de barras (MAE y olvido) para std y accurate
# ----------------------------------------------------------------------

def plot_mae_bars(winners_df: pd.DataFrame, preset_name: str):
    w = winners_df[winners_df["preset"] == preset_name].copy()
    if w.empty:
        print(f"[WARN] No hay winners para preset={preset_name}")
        return

    # Orden de métodos incluyendo rehearsal
    methods_order = ["std", "ewc", "rehearsal", "sa-snn", "as-snn", "sca-snn"]
    methods_present = [m for m in methods_order if m in w["method_base"].unique()]

    w = w.set_index("method_base").loc[methods_present].reset_index()

    plt.figure(figsize=(6, 4))
    x = range(len(w))
    width = 0.35

    plt.bar([i - width / 2 for i in x], w["c1_final_mae"], width=width, label="Circuito 1")
    plt.bar([i + width / 2 for i in x], w["c2_final_mae"], width=width, label="Circuito 2")

    plt.xticks(list(x), w["method_base"], rotation=45)
    plt.ylabel("MAE final")
    plt.title(f"MAE final por método (preset {preset_name})")
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.4)
    plt.tight_layout()
    plt.show()


def plot_forget_bars(winners_df: pd.DataFrame, preset_name: str):
    w = winners_df[winners_df["preset"] == preset_name].copy()
    if w.empty:
        print(f"[WARN] No hay winners para preset={preset_name}")
        return

    methods_order = ["std", "ewc", "rehearsal", "sa-snn", "as-snn", "sca-snn"]
    methods_present = [m for m in methods_order if m in w["method_base"].unique()]
    w = w.set_index("method_base").loc[methods_present].reset_index()

    plt.figure(figsize=(6, 4))
    plt.bar(w["method_base"], w["avg_forget_rel"])
    plt.ylabel("Olvido relativo medio")
    plt.title(f"Olvido relativo medio por método (preset {preset_name})")
    plt.grid(axis="y", linestyle="--", alpha=0.4)
    plt.tight_layout()
    plt.show()


# --- Figuras para std ---
plot_mae_bars(winners_compact, "std")
plot_forget_bars(winners_compact, "std")

# --- Figuras para accurate (incluye rehearsal) ---
plot_mae_bars(winners_compact, "accurate")
plot_forget_bars(winners_compact, "accurate")

# ----------------------------------------------------------------------
# 5. CO₂e vs MAE: figuras pensadas para la memoria
#    (a) Winners solamente  → limpio y legible
#    (b) Opcional: nube de runs filtrando outliers de energía
# ----------------------------------------------------------------------
try:
    df_std_sel = ensure_mae_aliases(load_csv(STD_SELECTION))
    df_acc_sel = ensure_mae_aliases(load_csv(ACC_SELECTION))
    df_all = pd.concat([df_std_sel, df_acc_sel], ignore_index=True)
    print("[INFO] CO₂e: usando selection_table (comparables)")
except Exception as e:
    print("[WARN] No se pudo leer selection_table, uso results_table_fromdisk:", e)
    df_all = pd.concat([df_std_all, df_acc_all], ignore_index=True)

# Normaliza columnas clave
df_all = ensure_mae_aliases(df_all)
if "method_base" in df_all.columns:
    df_all["method_base"] = df_all["method_base"].astype(str).apply(canonical_method)
else:
    df_all["method_base"] = df_all["method"].astype(str).apply(canonical_method)

# ---------- (a) Figura 1: solo winners (recomendada para la memoria) ----------
# winners_compact ya viene de winners_std + winners_acc
w = winners_compact.copy()
w_valid = w.dropna(subset=["emissions_kg"])
w = ensure_mae_aliases(w)
if "method_base" in w.columns:
    w["method_base"] = w["method_base"].astype(str).apply(canonical_method)
else:
    w["method_base"] = w["method"].astype(str).apply(canonical_method)

plt.figure(figsize=(7, 5))

markers = {"std": "o", "accurate": "^"}
for preset in ["std", "accurate"]:
    sub = w[w["preset"] == preset]
    if sub.empty:
        continue
    plt.scatter(
        sub["emissions_kg"],
        sub["c2_final_mae"],
        marker=markers.get(preset, "o"),
        label=f"preset={preset}",
        alpha=0.9,
    )
    # Etiquetamos cada punto con el método para que se entienda
    for _, row in sub.iterrows():
        plt.text(
            row["emissions_kg"],
            row["c2_final_mae"],
            f" {row['method_base']}",
            fontsize=8,
            va="center",
        )

plt.xlabel("Emisiones CO₂e (kg)")
plt.ylabel("MAE final Circuito 2")
plt.title("CO₂e vs MAE final (runs ganadores por método y preset)")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()


# ---------- (b) Figura 2 (opcional): nube completa con filtro de cuantiles ----------
def filter_energy_quantiles(df: pd.DataFrame,
                            group_cols=("preset", "method_base"),
                            q_low: float = 0.1,
                            q_high: float = 0.9,
                            min_abs: float = 1e-3) -> pd.DataFrame:
    """
    Mantiene solo los runs con emisiones entre los cuantiles [q_low, q_high]
    de su grupo (preset, método). Además exige emissions_kg > min_abs.

    Sirve para quitar:
      - runs con emisiones ~0 por errores de logging,
      - runs con emisiones desproporcionadas por debug/cuelgues.

    Úsalo solo para figuras exploratorias de CO₂e (no para las tablas base).
    """
    df = df.copy()
    df = df[df["emissions_kg"] > min_abs]

    if not group_cols:
        return df

    mask_keep = pd.Series(False, index=df.index)
    for _, g in df.groupby(list(group_cols)):
        v = g["emissions_kg"].astype(float)
        lo = v.quantile(q_low)
        hi = v.quantile(q_high)
        keep_g = (v >= lo) & (v <= hi)
        mask_keep.loc[g.index] = keep_g

    return df[mask_keep]


df_co2 = filter_energy_quantiles(df_all,
                                 group_cols=("preset", "method_base"),
                                 q_low=0.1,
                                 q_high=0.9,
                                 min_abs=1e-3)

print(f"[INFO] Runs energía: antes={len(df_all)} después filtro={len(df_co2)}")

plt.figure(figsize=(7, 5))
for preset in ["std", "accurate"]:
    sub = df_co2[df_co2["preset"] == preset]
    if sub.empty:
        continue
    plt.scatter(
        sub["emissions_kg"],
        sub["c2_final_mae"],
        alpha=0.4,
        label=f"preset={preset}",
    )

# Resaltamos los winners para referencia
for _, row in w.iterrows():
    plt.scatter(
        row["emissions_kg"],
        row["c2_final_mae"],
        s=80,
        facecolors="none",
        edgecolors="k",
    )

plt.xlabel("Emisiones CO₂e (kg)")
plt.ylabel("MAE final Circuito 2")
plt.title("CO₂e vs MAE final (runs comparables filtrados; winners resaltados)")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()



# ----------------------------------------------------------------------
# 6. Curvas por época (loss_curves.csv en task_1/2)
# ----------------------------------------------------------------------

def find_loss_curves_for_run(run_dir_rel: str) -> Dict[str, Optional[pd.DataFrame]]:
    """
    run_dir_rel viene de winners_compact['run_dir'], relativo a OUTPUTS_ROOT.
    Devuelve:
      {
        "circuito1": df_loss_curves_o_None,
        "circuito2": df_loss_curves_o_None,
      }
    """
    base = OUTPUTS_ROOT / run_dir_rel
    out = {}
    for task_idx, task_name in [(1, "circuito1"), (2, "circuito2")]:
        p = base / f"task_{task_idx}_{task_name}" / "loss_curves.csv"
        if p.exists():
            try:
                out[task_name] = pd.read_csv(p)
            except Exception:
                out[task_name] = None
        else:
            out[task_name] = None
    return out


def plot_learning_curves(
    winners_df: pd.DataFrame,
    preset_name: str,
    methods_to_compare: List[str],
):
    """
    Dibuja curvas val_mae por época para un conjunto de methods_to_compare
    usando los runs 'winners' del preset indicado.
    """
    w = winners_df[winners_df["preset"] == preset_name].copy()
    if w.empty:
        print(f"[WARN] No hay winners para preset={preset_name}")
        return

    w = w.copy()
    if "method_base" in w.columns:
        w["method_base"] = w["method_base"].astype(str).apply(canonical_method)

    w = w.set_index("method_base")

    plt.figure(figsize=(8, 5))
    algo_ploteado = False

    for mb in methods_to_compare:
        if mb not in w.index:
            print(f"[WARN] Método {mb} no presente en winners del preset {preset_name}, se omite")
            continue

        run_dir = w.loc[mb, "run_dir"]
        lc = find_loss_curves_for_run(run_dir)
        print(f"[INFO] {preset_name} / {mb} → {run_dir}")

        for task_name, df_loss in lc.items():
            if df_loss is None or df_loss.empty:
                print(f"  [WARN] Sin loss_curves para {task_name}")
                continue

            # Detectamos columna de MAE
            mae_col = None
            for cand in ["val_mae", "mae_val", "val_mae_epoch", "mae"]:
                if cand in df_loss.columns:
                    mae_col = cand
                    break
            if mae_col is None:
                print(f"  [WARN] Sin columna MAE reconocible en {task_name}, columnas: {list(df_loss.columns)}")
                continue

            if "epoch" not in df_loss.columns:
                print(f"  [WARN] Sin columna epoch en {task_name}, columnas: {list(df_loss.columns)}")
                continue

            label = f"{mb} – {task_name}"
            plt.plot(df_loss["epoch"], df_loss[mae_col], label=label)
            algo_ploteado = True

    if algo_ploteado:
        plt.xlabel("Época")
        plt.ylabel("MAE validación")
        plt.title(f"Curvas de aprendizaje/olvido (preset {preset_name})")
        plt.legend()
        plt.grid(True, linestyle="--", alpha=0.4)
        plt.tight_layout()
        plt.show()
    else:
        print("[WARN] No se ha podido generar la figura (faltan loss_curves válidos).")


# Ejemplo: comparar EWC, SCA-SNN y Rehearsal en preset accurate
methods_to_compare_acc = ["ewc", "rehearsal", "sa-snn", "as-snn", "sca-snn"]
plot_learning_curves(winners_compact, preset_name="accurate", methods_to_compare=methods_to_compare_acc)

# Ejemplo: comparar EWC y SCA-SNN en preset std (si los tienes)
methods_to_compare_std = ["ewc", "rehearsal", "sa-snn", "as-snn", "sca-snn"]
plot_learning_curves(winners_compact, preset_name="std", methods_to_compare=methods_to_compare_std)

# ----------------------------------------------------------------------
# 7. Análisis HPO agregado por método (tabla_hpo_por_metodo_agg_seeds.csv)
# ----------------------------------------------------------------------
HPO_AGG_CSV = STD_DIR.parent / "tabla_hpo_por_metodo_agg_seeds.csv"
df_hpo_agg = load_csv(HPO_AGG_CSV)

print(f"[INFO] HPO_AGG_CSV: {HPO_AGG_CSV}")
print(f"[INFO] df_hpo_agg: {df_hpo_agg.shape}")
print(f"[INFO] métodos base en HPO agg: {sorted(df_hpo_agg['method_base'].unique())}")

# Qué hiperparámetros queremos enseñar por método
HPO_PARAMS = {
    "as-snn": [
        "mp_gamma_ratio",
        "mp_lambda_a",
        "mp_ema",
        "mp_penalty_mode",
        "mp_do_synaptic_scaling",
        "mp_measure_at",
    ],
    "sa-snn": [
        "mp_attach_to",
        "mp_k",
        "mp_tau",
        "mp_vt_scale",
        "mp_th_min",
        "mp_th_max",
        "mp_p",
        "mp_assume_binary_spikes",
        "mp_flatten_spatial",
    ],
    "sca-snn": [
        "mp_attach_to",
        "mp_flatten_spatial",
        "mp_num_bins",
        "mp_anchor_batches",
        "mp_beta",
        "mp_bias",
        "mp_soft_mask_temp",
    ],
    "rehearsal": [
        "mp_buffer_size",
        "mp_replay_ratio",
    ],
    "ewc": [
        "mp_lambd",
        "mp_fisher_batches",
    ],
}


def show_hpo(method_base: str, top_k: int = 5):
    """Muestra hasta top_k configuraciones agregadas por método_base,
    sin duplicar filas con los mismos hiperparámetros.
    """
    d = df_hpo_agg[df_hpo_agg["method_base"] == method_base].copy()
    if d.empty:
        print(f"\n[WARN] No hay filas HPO agg para method_base={method_base}")
        return

    # Orden por métrica principal
    if "c2_mae_mean" in d.columns:
        d = d.sort_values("c2_mae_mean")
    elif "c2_final_mae" in d.columns:
        d = d.sort_values("c2_final_mae")

    d = d.head(top_k)

    # Columnas base: solo las que existan en el CSV (para evitar KeyError)
    base_candidates = [
        "preset",
        "method_base",
        "method",
        "n_runs",
        "c2_mae_mean",
        "c2_final_mae",
        "forget_mean",
        "avg_forget_rel",
        "co2_mean",
        "co2",
    ]
    cols_base = [c for c in base_candidates if c in d.columns]

    # Columnas mp_ (hiperparámetros)
    mp_cols_all = [c for c in d.columns if c.startswith("mp_")]

    preferred = HPO_PARAMS.get(method_base, [])
    mp_cols = [c for c in preferred if c in d.columns]

    if not mp_cols:
        mp_cols = mp_cols_all

    cols = cols_base + mp_cols

    print(f"\n### HPO (agg) para {method_base} (top-{len(d)}) ###")
    display(d[cols])


# Llamadas de ejemplo para inspeccionar cada método
show_hpo("ewc")
show_hpo("rehearsal")
show_hpo("sa-snn")
show_hpo("as-snn")
show_hpo("sca-snn")

import matplotlib.pyplot as plt
import numpy as np

def plot_mae_forget_grid(winners_df, preset: str, savepath: str | None = None):
    """
    Dibuja una figura 1x2:
      - Izquierda: barras de c2_final_mae por método.
      - Derecha: barras de avg_forget_rel por método.
    Usando SOLO los runs ganadores (winners_df) para un preset dado.
    """
    df = winners_df[winners_df["preset"] == preset].copy()
    if df.empty:
        print(f"[WARN] No hay filas en winners para preset={preset}")
        return

    # Columnas robustas (por si en winners_compact tienes 'method' en vez de 'method_base', etc.)
    method_col = "method_base" if "method_base" in df.columns else "method"
    mae_col    = "c2_final_mae"
    forget_col = "avg_forget_rel" if "avg_forget_rel" in df.columns else "forget"

    # Ordenar por MAE para que el gráfico tenga sentido visual
    df = df.sort_values(mae_col)
    methods = df[method_col].tolist()
    x = np.arange(len(methods))

    maes    = df[mae_col].values
    forgets = df[forget_col].values

    fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharex=True)
    ax_mae, ax_forget = axes

    # --- subplot izquierda: MAE ---
    ax_mae.bar(x, maes)
    ax_mae.set_title(f"{preset} – MAE final en circuito 2")
    ax_mae.set_ylabel("c2_final_mae")
    ax_mae.set_xticks(x)
    ax_mae.set_xticklabels(methods, rotation=45, ha="right")

    # --- subplot derecha: olvido ---
    ax_forget.bar(x, forgets)
    ax_forget.set_title(f"{preset} – Olvido relativo en circuito 1")
    ax_forget.set_ylabel("avg_forget_rel")
    ax_forget.set_xticks(x)
    ax_forget.set_xticklabels(methods, rotation=45, ha="right")

    fig.suptitle(f"Resumen de métodos – preset {preset}", y=1.02, fontsize=12)
    fig.tight_layout()

    if savepath is not None:
        fig.savefig(savepath, dpi=300, bbox_inches="tight")

    plt.show()

plot_mae_forget_grid(winners_compact, "std")
plot_mae_forget_grid(winners_compact, "accurate")


# ----------------------------------------------------------------------
# Figura 4.3 — Relación entre MAE (c2_final_mae) y emisiones_kg (CO₂e)
# Usamos las configuraciones ganadoras por método y preset
# ----------------------------------------------------------------------

import matplotlib.pyplot as plt

def plot_co2_vs_mae(winners_df):
    """
    winners_df: DataFrame con al menos las columnas
      ['preset', 'method_base', 'c2_final_mae', 'emissions_kg']
    """
    df = winners_df.copy()

    # Nos quedamos solo con las filas que tienen CO₂e (para evitar NaN, p.ej. Rehearsal accurate)
    df = df[~df["emissions_kg"].isna()].reset_index(drop=True)

    presets = ["std", "accurate"]

    fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
    fig.suptitle("Relación MAE vs emisiones de CO₂e (configuraciones ganadoras)")

    for ax, preset in zip(axes, presets):
        d = df[df["preset"] == preset]

        if d.empty:
            ax.set_title(f"preset = {preset} (sin datos)")
            ax.set_xlabel("Emisiones de CO₂e [kg]")
            continue

        # Scatter por método
        for _, row in d.iterrows():
            x = row["emissions_kg"]
            y = row["c2_final_mae"]
            label = row["method_base"]

            ax.scatter(x, y)
            # Etiqueta pequeña al lado del punto
            ax.text(
                x,
                y,
                label,
                fontsize=9,
                ha="left",
                va="bottom"
            )

        ax.set_title(f"preset = {preset}")
        ax.set_xlabel("Emisiones de CO₂e [kg]")

    axes[0].set_ylabel("MAE final circuito 2")

    plt.tight_layout()
    plt.show()


# Llamada con tu DataFrame de ganadores
# Si se llama distinto, cambia 'winners_compact' por el nombre real.
plot_co2_vs_mae(winners_compact)

# ----------------------------------------------------------------------
# 6. Curvas por época (Figura 4.4) usando loss_curves.csv en task_1/2
# ----------------------------------------------------------------------

def find_loss_curves_for_run(run_dir_rel: str) -> Dict[str, Optional[pd.DataFrame]]:
    """
    run_dir_rel viene de winners_compact['run_dir'], relativo a OUTPUTS_ROOT.
    Devuelve:
      {
        "circuito1": df_loss_curves_o_None,
        "circuito2": df_loss_curves_o_None,
      }
    """
    base = OUTPUTS_ROOT / run_dir_rel
    out = {}
    for task_idx, task_name in [(1, "circuito1"), (2, "circuito2")]:
        p = base / f"task_{task_idx}_{task_name}" / "loss_curves.csv"
        if p.exists():
            try:
                out[task_name] = pd.read_csv(p)
            except Exception:
                out[task_name] = None
        else:
            out[task_name] = None
    return out


def plot_learning_curves(
    winners_df: pd.DataFrame,
    preset_name: str,
    methods_to_compare: List[str],
    only_circuito2: bool = True,
):
    """
    Dibuja curvas de MAE de validación por época para los métodos indicados,
    usando los runs 'winners' del preset indicado.

    - winners_df: normalmente winners_compact (con columnas 'preset','method_base','run_dir').
    - preset_name: 'std' o 'accurate'.
    - methods_to_compare: lista de métodos base (ewc, rehearsal, sa-snn, as-snn, sca-snn).
    - only_circuito2: si True, solo plotea las curvas de circuito2 (tarea B).
    """
    w = winners_df[winners_df["preset"] == preset_name].copy()
    if w.empty:
        print(f"[WARN] No hay winners para preset={preset_name}")
        return

    # Normalizamos nombres de método si tienes canonical_method definido
    if "method_base" in w.columns:
        w["method_base"] = w["method_base"].astype(str)
        if "canonical_method" in globals():
            w["method_base"] = w["method_base"].apply(canonical_method)

    w = w.set_index("method_base")

    plt.figure(figsize=(8, 5))
    algo_ploteado = False

    for mb in methods_to_compare:
        if mb not in w.index:
            print(f"[WARN] Método {mb} no presente en winners del preset {preset_name}, se omite")
            continue

        run_dir = w.loc[mb, "run_dir"]
        lc = find_loss_curves_for_run(run_dir)
        print(f"[INFO] {preset_name} / {mb} → {run_dir}")

        for task_name, df_loss in lc.items():
            if df_loss is None or df_loss.empty:
                print(f"  [WARN] Sin loss_curves para {task_name}")
                continue

            # Si solo queremos B (circuito2), filtramos aquí
            if only_circuito2 and task_name != "circuito2":
                continue

            # Detectamos columna de MAE
            mae_col = None
            for cand in ["val_mae", "mae_val", "val_mae_epoch", "mae"]:
                if cand in df_loss.columns:
                    mae_col = cand
                    break

            if mae_col is None:
                print(f"  [WARN] Sin columna MAE reconocible en {task_name}, columnas: {list(df_loss.columns)}")
                continue

            if "epoch" not in df_loss.columns:
                print(f"  [WARN] Sin columna epoch en {task_name}, columnas: {list(df_loss.columns)}")
                continue

            # Etiqueta de la curva
            if only_circuito2:
                label = mb  # ya sabemos que es circuito2
            else:
                label = f"{mb} – {task_name}"

            plt.plot(df_loss["epoch"], df_loss[mae_col], label=label)
            algo_ploteado = True

    if algo_ploteado:
        plt.xlabel("Época")
        plt.ylabel("MAE validación")
        title_task = " (circuito 2)" if only_circuito2 else ""
        plt.title(f"Curvas de aprendizaje/olvido – preset {preset_name}{title_task}")
        plt.legend()
        plt.grid(True, linestyle="--", alpha=0.4)
        plt.tight_layout()

        # Si quieres guardar la figura para Word:
        # plt.savefig(f"fig_4_4_learning_curves_{preset_name}.png", dpi=300)

        plt.show()
    else:
        print("[WARN] No se ha podido generar la figura (faltan loss_curves válidos).")


# ----------------------------------------------------------------------
# Llamadas para generar la Figura 4.4
# ----------------------------------------------------------------------

methods_to_compare_acc = ["ewc", "rehearsal", "sa-snn", "as-snn", "sca-snn"]
plot_learning_curves(
    winners_compact,
    preset_name="accurate",
    methods_to_compare=methods_to_compare_acc,
    only_circuito2=True,   # solo B, que es lo que se comenta en el texto
)

# Si quisieras una figura análoga para std, la puedes generar así:
methods_to_compare_std = ["ewc", "rehearsal", "sa-snn", "as-snn", "sca-snn"]
plot_learning_curves(
    winners_compact,
    preset_name="std",
    methods_to_compare=methods_to_compare_std,
    only_circuito2=True,
)

# ----------------------------------------------------------------------
# 6. Curvas por época (Figura 4.4) usando loss_curves.csv en task_1/2
# ----------------------------------------------------------------------

def find_loss_curves_for_run(run_dir_rel: str) -> Dict[str, Optional[pd.DataFrame]]:
    """
    run_dir_rel viene de winners_compact['run_dir'], relativo a OUTPUTS_ROOT.
    Devuelve:
      {
        "circuito1": df_loss_curves_o_None,
        "circuito2": df_loss_curves_o_None,
      }
    """
    base = OUTPUTS_ROOT / run_dir_rel
    out = {}
    for task_idx, task_name in [(1, "circuito1"), (2, "circuito2")]:
        p = base / f"task_{task_idx}_{task_name}" / "loss_curves.csv"
        if p.exists():
            try:
                out[task_name] = pd.read_csv(p)
            except Exception:
                out[task_name] = None
        else:
            out[task_name] = None
    return out


def _plot_learning_curves_single_preset(
    winners_df: pd.DataFrame,
    preset_name: str,
    methods_to_compare: List[str],
    ax,
    only_circuito2: bool = True,
):
    """
    Lógica interna: dibuja las curvas de un preset concreto sobre el eje `ax`.
    """
    w = winners_df[winners_df["preset"] == preset_name].copy()
    if w.empty:
        print(f"[WARN] No hay winners para preset={preset_name}")
        return False

    # Normalizamos nombres de método si tienes canonical_method definido
    if "method_base" in w.columns:
        w["method_base"] = w["method_base"].astype(str)
        if "canonical_method" in globals():
            w["method_base"] = w["method_base"].apply(canonical_method)

    w = w.set_index("method_base")

    algo_ploteado = False

    for mb in methods_to_compare:
        if mb not in w.index:
            print(f"[WARN] Método {mb} no presente en winners del preset {preset_name}, se omite")
            continue

        run_dir = w.loc[mb, "run_dir"]
        lc = find_loss_curves_for_run(run_dir)
        print(f"[INFO] {preset_name} / {mb} → {run_dir}")

        for task_name, df_loss in lc.items():
            if df_loss is None or df_loss.empty:
                print(f"  [WARN] Sin loss_curves para {task_name}")
                continue

            # Si solo queremos B (circuito2), filtramos aquí
            if only_circuito2 and task_name != "circuito2":
                continue

            # Detectamos columna de MAE
            mae_col = None
            for cand in ["val_mae", "mae_val", "val_mae_epoch", "mae"]:
                if cand in df_loss.columns:
                    mae_col = cand
                    break

            if mae_col is None:
                print(f"  [WARN] Sin columna MAE reconocible en {task_name}, columnas: {list(df_loss.columns)}")
                continue

            if "epoch" not in df_loss.columns:
                print(f"  [WARN] Sin columna epoch en {task_name}, columnas: {list(df_loss.columns)}")
                continue

            # Etiqueta de la curva (solo método, ya sabemos que es circuito2)
            label = mb if only_circuito2 else f"{mb} – {task_name}"

            ax.plot(df_loss["epoch"], df_loss[mae_col], label=label)
            algo_ploteado = True

    if algo_ploteado:
        ax.set_xlabel("Época")
        ax.set_ylabel("MAE validación")
        ax.set_title(f"preset {preset_name} – circuito 2")
        ax.grid(True, linestyle="--", alpha=0.4)

    return algo_ploteado


def plot_learning_curves_grid(
    winners_df: pd.DataFrame,
    methods_to_compare: List[str],
    only_circuito2: bool = True,
):
    """
    Genera una figura con dos subgráficos:
      - Izquierda: curvas de aprendizaje para preset std
      - Derecha: curvas de aprendizaje para preset accurate
    """
    fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

    plotted_std = _plot_learning_curves_single_preset(
        winners_df,
        preset_name="std",
        methods_to_compare=methods_to_compare,
        ax=axes[0],
        only_circuito2=only_circuito2,
    )

    plotted_acc = _plot_learning_curves_single_preset(
        winners_df,
        preset_name="accurate",
        methods_to_compare=methods_to_compare,
        ax=axes[1],
        only_circuito2=only_circuito2,
    )

    # Leyenda común si hay algo ploteado
    if plotted_std or plotted_acc:
        handles, labels = axes[1].get_legend_handles_labels()

        # Título un poco por encima de los subplots, no tan arriba del todo
        fig.suptitle(
            "Curvas de MAE de validación en el circuito 2 – métodos ganadores",
            y=0.88,
        )

        # Leyenda cerca de los ejes inferiores (no tan separada)
        fig.legend(
            handles,
            labels,
            loc="lower center",
            bbox_to_anchor=(0.5, 0.03), 
            ncol=len(labels),
        )

        # Dejamos hueco ARRIBA (para el título) y ABAJO (para la leyenda)
        fig.tight_layout(rect=[0, 0.08, 1, 0.92])
        #         left  bottom  right  top

        plt.show()
    else:
        print("[WARN] No se ha podido generar la figura (faltan loss_curves válidos).")


# ----------------------------------------------------------------------
# Llamada para generar la Figura 4.4 (grid std + accurate)
# ----------------------------------------------------------------------

methods_to_compare = ["ewc", "rehearsal", "sa-snn", "as-snn", "sca-snn"]
plot_learning_curves_grid(
    winners_compact,
    methods_to_compare=methods_to_compare,
    only_circuito2=True,
)
```
